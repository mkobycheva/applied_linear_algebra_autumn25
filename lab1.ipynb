## Importing libraries + creating 2-D picture

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
```

```python
with open("two.csv") as f:
    lines = [line.replace(',', '.') for line in f]  # fix commas in decimals
    two = np.loadtxt(lines, delimiter=';', skiprows=1)

two[:,1] = max(two[:,1]) - two[:,1]
plt.plot(two[:,0], two[:,1])  # replace `batman` with any array
plt.axis('equal')
plt.show()
```

# Task 1

```python
def plotting(graph):
    plt.plot(graph[:,0], graph[:,1])
    plt.axis('equal')
    plt.show()
```

```python
def stretch(X, a, b):
    matrix = np.array([
        [a, 0],
        [0, b]
    ])
    output = X @ matrix
    return output

output = stretch(two, 1, 2)
plotting(output)
print(output)
```

```python
def shear(X, a, b):
    matrix = np.array([
        [1, a],
        [b, 1]
    ])
    output = X @ matrix
    return output

output = shear(two, 0.5, 0)
plotting(output)
print(output)
```

```python
def reflection(X, a, b):
    matrix = (1/(np.sqrt(a) + np.sqrt(b))) * np.array([
        [np.sqrt(a) - np.sqrt(b), 2 * a * b],
        [2 * a * b, np.sqrt(b) - np.sqrt(a)]
    ])
    output = X @ matrix
    return output

output = reflection(two, 1, 2)
plotting(output)
print(output)
```

```python
def rotation(X, t):
    matrix = np.array([
        [np.cos(t), -np.sin(t)],
        [np.sin(t), np.cos(t)]
    ])
    output = X @ matrix
    return output

output = rotation(two, 3*np.pi/4)
plotting(output)
print(output)
```

# 3-D Picture

```python
def read_off(filename: str):
    with open(filename, 'r') as f:
        if "OFF" != f.readline().strip():
            raise ValueError

        n_verts, n_faces, _ = map(int, f.readline().strip().split())

        verts = [list(map(float, f.readline().strip().split())) for _ in range(n_verts)]
        faces = [list(map(int, f.readline().strip().split()[1:])) for _ in range(n_faces)]
    return np.array(verts), faces

vertices, faces = read_off("/Users/masha/PycharmProjects/linear_algebra/3dpic/ModelNet40/lamp/test/lamp_0138.off")
print(vertices.shape)
print(faces[:3])
print(vertices)
```

```python
def plot_off(vertices, faces):
    fig = plt.figure(figsize=(8, 8))
    ax = fig.add_subplot(111, projection='3d')

    mesh = Poly3DCollection([vertices[face] for face in faces],
                        alpha=0.3, edgecolors='k', linewidths=0.1)
    ax.add_collection3d(mesh)

    ax.scatter(vertices[:, 0], vertices[:, 1], vertices[:, 2], s=2, c='r')

    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')

    ax.auto_scale_xyz(vertices[:, 0], vertices[:, 1], vertices[:, 2])

    plt.show()

plot_off(vertices, faces)
```

# Task 3

```python
def rotate_xy(X, t):
    matrix = np.array([
        [np.cos(t), -np.sin(t), 0],
        [np.sin(t), np.cos(t), 0],
        [0, 0, 1]
    ])
    output = X @ matrix
    return output

new_vertices = rotate_xy(vertices, np.pi/4)
plot_off(new_vertices, faces)
print(new_vertices)
```

```python
def rotate_yz(X, t):
    matrix = np.array([
        [1, 0, 0],
        [0, np.cos(t), -np.sin(t)],
        [0, np.sin(t), np.cos(t)]
    ])
    output = X @ matrix
    return output

new_vertices = rotate_yz(vertices, np.pi/4)
plot_off(new_vertices, faces)
print(new_vertices)
```

```python
def rotate_xz(X, t):
    matrix = np.array([
        [np.cos(t), 0, -np.sin(t)],
        [0, 1, 0],
        [np.sin(t), 0, np.cos(t)]
    ])
    output = X @ matrix
    return output


new_vertices = rotate_xz(vertices, np.pi / 4)
plot_off(new_vertices, faces)
print(new_vertices)
```
